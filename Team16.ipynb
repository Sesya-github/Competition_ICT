```python
from geopy import distance
#Finding the distance between origin and destination
distances_km = []
for row in data.itertuples(index=False):
   distances_km.append(distance.distance(row.Org_lat_lon, row.Des_lat_lon).km)
data['Org_Dest_distance'] = distances_km
```

```python
#Finding the registered states of vehicles
data['vehicle_states'] = data.vehicle_no.astype(str).str[:2]
data['vehicle_states'].unique()
```

```python
#Replace tn,hr with TN,HR
data['vehicle_states']=data['vehicle_states'].replace(('tn', 'hr'), ('TN', 'HR'))
data['vehicle_states'].unique()
```

```python
#Finding Originating states from Originating location
data['Origin_states'] = data['Origin_Location'].str.split(',').apply(lambda x: x[-1])
data['Origin_states'].unique()
```

```python
for i in data.index:
  if data['Origin_states'][i]==' India':
    data['Origin_states'][i]=data['Origin_Location'][i].split(',')[-2]#Taking second last string to fill the Origin_state column
data['Origin_states'].unique()
```

```python

data['Origin_states']=data['Origin_states'].replace(('TAMIL NADU', 'PONDICHERRY', 'KARNATAKA', 'RAJASTHAN',
       'UTTAR PRADESH', 'HARYANA','Haryana','MAHARASHTRA', 'GUJARAT'),(' Tamil Nadu',' Pondicherry',' Karnataka',
    ' Rajasthan',' Uttar Pradesh',' Haryana',' Haryana',' Maharashtra',' Gujarat'))
data['Origin_states'].unique()
                                                    
```

```python
data['Origin_states']=data['Origin_states'].replace(('Sedarapet','Karnataka 562114'),(' Puducherry',' Karnataka'))
data['Origin_states'].unique()
```

```python
data['Dest_states'] = data['Destination_Location'].str.split(',').apply(lambda x: x[-1])
data['Dest_states'].unique()
```

```python
#Finding the Dest_states from the Destination_Location
for i in data.index:
  if data['Dest_states'][i]==' India':
    data['Dest_states'][i]=data['Destination_Location'][i].split(',')[-2]
data['Dest_states'].unique()
```

```python
data['Dest_states']=data['Dest_states'].replace(('KARNATAKA', 'TAMIL NADU', 'PONDICHERRY', 'RAJASTHAN', 
       'MAHARASHTRA', 'GUJARAT','JHARKHAND','Jharkhand', 'Haryana', 'HARYANA', 'Chhattisgarh','Chattisgarh',
        'HIMACHAL PRADESH','Goa','West Bengal','Sikkim','Bihar','Karnataka 560300'),(' Karnataka',' Tamil Nadu',
      ' Pondicherry',' Rajasthan',' Maharashtra',' Gujarat',' Jharkhand',' Jharkhand',' Haryana',' Haryana',
    ' Chattisgarh',' Chattisgarh',' Himachal Pradesh',' Goa',' West Bengal',' Sikkim',' Bihar',' Karnataka'))
data['Dest_states'].unique()
```

# Missing value treatment

```python
#Finding the percentage of missing values in each feature
for col in data.columns:
    if data[col].isna().sum()>0:
        print(col, data[col].isna().mean().round(4)*100)
```

```python
#Name 'Unkown' for null values in Driver_Name
data['Driver_Name']=data['Driver_Name'].fillna('Unknown')

#Impute transportation distence with mean value
data['TRANSPORTATION_DISTANCE_IN_KM']= data['TRANSPORTATION_DISTANCE_IN_KM'].rolling(min_periods=1, center=True, window=12).mean()

#Name 'Unkown' for null values in vehicleTtype
data['vehicleType']=data['vehicleType'].fillna('Unknown')

#Filling pervious date for actual_eta
data['actual_eta']=data['actual_eta'].fillna(method='ffill')
```

```python
#Impute null values in trip_end_date
import datetime
import random
df_sub=data[data['trip_end_date'].isna()]
for i in df_sub.index:
  if df_sub['ontime/delay'][i]==0:
    df_sub['trip_end_date'][i]=df_sub['actual_eta'][i]
  else:
    df_sub['trip_end_date'][i]=df_sub['Planned_ETA'][i]-datetime.timedelta(random.randint(0,3))
data=pd.concat([data, df_sub])

#Since we are concatinationg those null in trip_end_date let's remove the duplicate rows
data.dropna(subset=['trip_end_date'], inplace=True)
```

```python
#Features like GpsProvider, Data_Ping_time, Current_Location, Curr_lat, Curr_lon having null values in same rows
#all the above mentioned features are dependent on each other and it's not feasible to impute those, so let's drop those rows

data.dropna(subset=['Current_Location'], inplace=True)
```

```python
data.dropna(subset=['OriginLocation_Code','DestinationLocation_Code'],inplace=True)
data['Driver_MobileNo'].fillna(0, inplace=True)#fill null values in Driver_MobileNo with zero 
#Actually there is no relevence to the Driver_MobileNo, but it is worth to know whether the Mobile no is available or not
#So we can give a 1 to Driver_MobileNo if it is available
data['Driver_MobileNo'].values[data['Driver_MobileNo'].values>0]=1
data.drop('Minimum_kms_to_be_covered_in_a_day',axis=1,inplace=True)#Drop this feature since it has no importance in our ML model
```
